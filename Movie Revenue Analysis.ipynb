## Exploratory Data Analysis

> Use Matplotlib to display your data analysis

### Which production companies released the most movies in the last 10 years? Display the top 5 production companies.

```python
production_companies = movies[['revenue_adj', 'production_companies','original_title']].sort_values(['revenue_adj'], ascending=False).reset_index(drop=True)
production_companies.head(5)
```

### What 5 movie genres grossed the highest all-time?

```python
genres_grossed_highest = movies[['revenue_adj', 'genres','imdb_id' ]].sort_values(['revenue_adj','genres'], ascending=False).reset_index(drop=True)
genres_grossed_highest.head(5)
```

### Who are the top 5 grossing directors?

```python
top5_grossing_directors= movies[['revenue_adj', 'director', 'imdb_id', ]].sort_values(['revenue_adj'], ascending=False).reset_index(drop=True)
top5_grossing_directors.head(10)
```

### Compare the revenue of the highest grossing movies of all time.

```python

movies[['original_title','genres','revenue_adj']].describe()
highest_gross = movies[['original_title','revenue_adj','imdb_id']].sort_values(['revenue_adj'], ascending=False).reset_index(drop=True)
highest_gross
highest_gross.head(5)
```
